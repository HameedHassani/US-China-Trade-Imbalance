```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px


data = pd.read_csv('forbes_billionaires_geo_3.csv')

data.head()
```

```python
# Top 10 Countries that have most billionaires

countries_with_most_billionaires = data['Citizenship'].value_counts()
top10 = countries_with_most_billionaires.head(10)
figure = top10.plot.barh()
figure.invert_yaxis()

plt.xlabel('Number of Billionaires')
plt.ylabel('Countries')
plt.title('Top 10 Countries with Most Billionaires')
```

```python
# Marital status of billionaires

status = ['Married', 'Divorced', 'Widowed', 'Single']
billionaires_status = data['Status'].isin(status)
data = data[billionaires_status]
four_status = data['Status'].value_counts()

explode = [0,0.1,0,0]
four_status.plot.pie(startangle = 50, explode = explode, shadow = True, autopct = '%1.1f%%',pctdistance=0.76)
plt.title('Billionaires Status')
```

```python
# Minimum Education Level of billionaires

phd = data['Education'].str.contains('Doctorate', na=False).value_counts()
Master = data['Education'].str.contains('Master', na=False).value_counts()
Bachelor = data['Education'].str.contains('Bachelor', na=False).value_counts()
Drop_Out = data['Education'].str.contains('Drop Out', na=False).value_counts()
High_School_Graduate = data['Education'].str.contains('High School Graduate', na=False).value_counts()

Edu_level = [phd, Master, Bachelor, Drop_Out, High_School_Graduate]

Edu_level
```

```python
# Minimum Education Level of billionaires
Edu_labels = ['Phd', 'Master','Bachelor', 'Drop Out','High School Graduate']
Edu_level_numbers = [62, 473, 1008, 78, 8]
plt.bar(Edu_labels, Edu_level_numbers)

plt.xlabel('Education Level (minimum)')
plt.ylabel('Number of Billionaires')
plt.title("Billionaires' Education Level")
plt.xticks(rotation=45)
```

```python
# billionaires age distribution

Billionaires_age = data['Age']

plt.hist(Billionaires_age, bins = 40)
plt.title("Billionaires' Age Distribution")
plt.xlabel('Age')
plt.ylabel('Number of Billionaires')
```

```python
# Top 10 Wealth Source of billionaires

wealth_source = data['Source'].value_counts().head(10)
wealth_source.plot.bar()

plt.title("Top 10 Sources of Wealth")
plt.xlabel('Industry')
plt.ylabel('Number of Billionaires')
```

```python
# Billionaires' Net Worth Distribution

net_worth = data['NetWorth']
plt.hist(net_worth, bins = 50, log = True)

plt.title("Billionaires' Net Worth Distribution")
plt.xlabel('Net Worth')
plt.ylabel('Number of Billionaires')
```
