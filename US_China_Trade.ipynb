```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('U.S.-China Trade.csv')
```

```python
  #  U.S.-China Trade (USCT)
usct = data[data['CTYNAME'] == 'China']
Year = usct['year']

  #  US import (USI) in billion $
USI = (usct['IYR'] * 0.001).round(2)

  #  US Export (USE) in billion $
USE = (usct['EYR'] * 0.001).round(2)

  # U.S.-China Total Trade (USCTT)
USCTT = USE +USI

  #US Trade Deficit with China 
Deficit = USE - USI

usct.head()

```

```python
plt.figure(figsize=(7,5))

width = 0.8

plt.bar(Year, USCTT, color='steelblue', width=width, label = 'Total Bilateral Trade')
plt.bar(Year, USI,color='orange',width=width, label = 'U.S. Import')
plt.bar(Year, USE,color='forestgreen', width=width , label = 'U.S. Export')
plt.bar(Year, Deficit, color='salmon', width=width , label = 'Trade Deficit')

plt.title('U.S.-China Merchandise Trade', fontsize = 15)
plt.xlabel('Year', fontsize = 14)
plt.xticks(Year, rotation=90)
plt.ylabel('Trade in Billion Dollar',  fontsize = 14)


plt.legend()
plt.grid(axis='y', alpha = 0.5)
plt.tight_layout()
plt.show()
```
